```python
def func(x):
    f=2*x-1.75*x*x+1.1*x*x*x-0.25*x*x*x*x
    return f
```

```python
R = 0.618033988
a = -2
b = 4
maxIter = 100
tolerence = 0.01
```

```python
x1=0
x2=0
xopt=0
for k in range(1,100):
    print("Interation",k)
    x1 = a + R*(b-a)
    x2 = b -R*(b-a)
    f1 = func(x1)
    print("f(x1) = ",f1)
    f2 = func(x2)
    print("f(x2) = ",f2)
    if f1 > f2:
        xopt=x1
        tolerence_calculated=abs((1-R)*(a-b)*100/abs(xopt))
        a = x2
        x2= x1
        f2= f1
        x1= a + R*(b-a)
        f1= func(x1)
    else:
        xopt=x2
        tolerence_calculated=abs((1-R)*(a-b)*100/abs(xopt))
        b = x1
        x1 = x2
        f1= f2
        x2= b -R*(b-a)
        f2= func(x2)
    print("x1 =", x1)
    print("x2 =", x2)
    print("tolerance =",tolerence_calculated,"%\n")
    if((x1-x2)<=tolerence):
        break
    else:
        continue
    
print("Point", x1, x2)
```
